##### Write expressions for finding all words in text6 that meet the conditions listed below. The result should be in the form of a list of words: ['word1', 'word2', ...].

1. Ending in *ise*
1. Containing the letter *z*
1. Containing the sequence of letters *pt*
1. Having all lowercase letters except for an initial capital (i.e., titlecase)

````python
# ending with ise
result = []
for w in text6:
    if w.endswith("ise"):
        result.append(w)
print(result)
````

````python
# Containing the letter z
result = []
for w in text6:
    if "z" in w:
        result.append(w)
print(result)
````

````python
# Containing the sequence of letters pt
result = []
for w in text6:
    if "pt" in w:
        result.append(w)
print(result)
````

````python
# titlecase
result = []
for w in text6:
    if w.istitle():
        result.append(w)
print(result)
````

##### Define a function called `vocab_size(text)` that has a single parameter for the text, and which returns the vocabulary size of the text.

````python
def vocab_size(text):
    return len(set(text))
````

##### Define a function `percent(word, text)` that calculates how often a given word occurs in a text, and expresses the result as a percentage.

````python
def percent(word, text):
    count = len(w for w in text if w == word)
    return 100 * count / len(text)
````
