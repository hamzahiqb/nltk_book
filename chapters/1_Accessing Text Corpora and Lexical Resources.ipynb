```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(
        round(num_chars / num_words),
        round(num_words / num_sents),
        round(num_words / num_vocab),
        fileid,
    )
```

*stylistics*: Semantic differences between genres

```python
from nltk.corpus import brown

news_text = brown.words(categories="news")
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ["can", "could", "may", "might", "must", "will"]
for m in modals:
    print(m + ":", fdist[m], end=" ")
```

```python
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)
genres = ["news", "religion", "hobbies", "science_fiction", "romance", "humor"]
modals = ["can", "could", "may", "might", "must", "will"]
cfd.tabulate(conditions=genres, samples=modals)
```

# Lexical Resources

A lexicon is a collection of words and/or phrases along with associated information such as part of speech and sense definition.

A lexical entry consists of a:
1. headword or lemma
1. additional information such as part of speech or sense definition

Two distinct words having the same spelling are called homonyms

## Wordlist Corpora

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)


unusual_words(nltk.corpus.gutenberg.words("austen-sense.txt"))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

```python
from nltk.corpus import stopwords

stopwords.words("english")
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words("english")
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)


content_fraction(nltk.corpus.reuters.words())
```

```python
puzzle_letters = nltk.FreqDist("egivrvonl")
obligatory = "r"
wordlist = nltk.corpus.words.words()
[
    w
    for w in wordlist
    if len(w) >= 6 and obligatory in w and nltk.FreqDist(w) <= puzzle_letters
]
```

## A Pronouncing Dictionary

```python
entries = nltk.corpus.cmudict.entries()
len(entries)
```

```python
for entry in entries[42371:42379]:
    print(entry)
```

```python
for word, pron in entries:
    if len(pron) == 3:
        ph1, ph2, ph3 = pron
        if ph1 == "P" and ph3 == "T":
            print(word, ph2, end=" ")
```

```python
syllable = ["N", "IH0", "K", "S"]
[word for word, pron in entries if pron[-4:] == syllable]
```

```python
[w for w, pron in entries if pron[-1] == "M" and w[-1] == "n"]
```

```python
sorted(set(w[:2] for w, pron in entries if pron[0] == "N" and w[0] != "n"))
```

```python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]
```

```python
[w for w, pron in entries if stress(pron) == ["0", "1", "0", "2", "0"]]
```

```python
p3 = [
    (pron[0] + "-" + pron[2], word)
    for (word, pron) in entries
    if pron[0] == "P" and len(pron) == 3
]

cfd = nltk.ConditionalFreqDist(p3)
for template in sorted(cfd.conditions()):
    if len(cfd[template]) > 10:
        words = sorted(cfd[template])
        wordstring = " ".join(words)
        print(template, wordstring[:70] + "...")
```

## More Lexical Relations

Hypernyms and Homonyms are lexical relations as they relate one synset to another. Others are

1. Meronyms: From one item to their components (E.g. tree to trunk, roots, etc), or
2. Holonyms: To the things they are contained in (e.g. trees are contained in a forest)
